```python
import qrcode
from PIL import Image

# Create QR code
img = qrcode.make('34.93506806162019, 8.570557891749553')
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)

# Add website data to the QR code
qr.add_data("https://treestory.duckdns.org/a00000000001/")
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white").convert('RGB')
```

```python
from matplotlib.pyplot import imshow
import numpy as np

# Scan the QR and check the stats of your tree!!
imshow(np.asarray(img))
```
