```python
import qrcode
from PIL import Image

# Create QR code
img = qrcode.make('34.93506806162019, 8.570557891749553')
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)

# Add website data to the QR code
qr.add_data("https://www.google.com/maps/place/34%C2%B056'06.3%22N+8%C2%B034'14.0%22E/@34.9350681,8.5700094,179m/data=!3m2!1e3!4b1!4m5!3m4!1s0x0:0x0!8m2!3d34.9350681!4d8.5705579")
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white").convert('RGB')
```

```python
from matplotlib.pyplot import imshow
import numpy as np

# QR code for location of individual tree
imshow(np.asarray(img))
```
